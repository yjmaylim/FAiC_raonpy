```
import pandas as pd
import numpy as np

data = pd.read_csv("coviddata.csv")
data

total = data["Total Cases"]
released = data["Released Cases"]
deaths = data["Deaths"]
```

```
%matplotlib inline  

import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도
print(mpl.__version__)
!apt-get update -qq
!apt-get install fonts-nanum* -qq
```

```
sys_font=fm.findSystemFonts()
nanum_font = [f for f in sys_font if 'Nanum' in f]
nanum_font
```

```
path = '/usr/share/fonts/truetype/nanum/NanumGothicEcoBold.ttf'  # 설치된 나눔글꼴중 원하는 녀석의 전체 경로를 가져오자
font_name = fm.FontProperties(fname=path, size=5).get_name()
fontprop = fm.FontProperties(fname=path, size=18)
fontprops = fm.FontProperties(fname=path, size=13)
plt.rc('font', family=font_name)
fm._rebuild()
```

```
import matplotlib.pyplot as plt
from datetime import datetime

date = pd.date_range(start="2020-03-10",end="2020-12-10")

plt.figure(figsize=(4, 8))
plt.plot(date, total, label='Total Cases', color="black")
plt.plot(date, released, label='Released Cases', color="blue")
plt.plot(date, deaths, label='Deaths', color="red")
plt.title('코로나19 확진자수',fontproperties=fontprop)  # 제목
plt.xlabel('날짜', fontproperties=fontprops)  # x축 이름
plt.xticks(rotation=45)
plt.ylabel('명(수)', fontproperties=fontprops)  # y축 이름
plt.legend()
plt.savefig("FAICcovid.png", dpi = 300, bbox_inches = "tight")
```
